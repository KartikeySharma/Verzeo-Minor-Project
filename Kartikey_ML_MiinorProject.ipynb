## ----------------------------------------------GRADES ASSESMENT---------------------------------------------

Consider the following grades for further analysis<br><br>
        Score Obtained   --------         Grade<br><br>
            85-100  ----------->               A+<br>
            80-84   ------------>               A<br>
            70-79   ------------>               B+<br>
            60-69   ------------>               B<br>
            55-59   ------------>               C+<br>
            45-54   ------------>               C<br>
            35-44   ------------>               D<br>
            Below 35  ------->               E<br>

```python
d['total marks']=d['math score']+d['reading score']+d['writing score']
d['average']=d['total marks']*100/300
```

```python
def gradingpolicy(scores):
    if scores >= 85 and scores <= 100:
        return 'A+'
    elif scores >= 80 and scores < 85:
        return 'A '
    elif scores >= 70 and scores < 80:
        return 'B+'
    elif scores >= 60 and scores < 70:
        return 'B'
    elif scores >= 55 and scores < 60:
        return 'C+'
    elif scores >= 45 and scores < 55:
        return 'C'
    elif scores >= 35 and scores < 45:
        return 'D'
    elif scores >= 0 and scores < 35:
        return 'E'
    
d['grades']=d['average'].apply(gradingpolicy)
```

```python
d.head(10)
```

```python
d.info()
```

### Distribution of Grades

```python
plt.figure(figsize=(8,8))
plt.title("Grades of Students",fontsize=15)
d['grades'].value_counts().plot.pie(autopct="%1.1f%%")
plt.show()
```

We observe that majority of the students have got B+ and B grade, which means that most students scored between 60 and 79 marks. More than 1/10th of the students scored Outstanding Result (A+) and around 9 % scored Excellent (A) grade.

## Relation between GRADES and other variables

#### GRADES VS GENDER

```python
plt.figure(figsize=(8,5))
plt.title("Gender vs Grade Obtained by Students",fontsize=15)
sns.countplot(data=d, hue="gender", x="grades")
plt.show()
```

We see that female students have outperformed male students in the better grades(A+,A,B+) while more male students have scored lower grades than female students.

#### GRADES VS LUNCH

```python
plt.figure(figsize=(8,5))
plt.title("Lunch Type vs Grade Obtained by Students",fontsize=15)
sns.countplot(data=d, hue="lunch", x="grades")
plt.show()
```

We observe that majority of the students securing good grades(A+,A,B+,B) had standard lunch, whereas for other grades (C,D,C+,E), majority of the students had free/reduced lunch

We can clearly see that more students who completed the test preparation course achieved excellent grades(A,A+)

#### GRADES VS PARENTAL LEVEL OF EDUCATION

```python
plt.figure(figsize=(15,8))
plt.title('Grades vs Parental level of Education',fontsize=15)
sns.countplot(x="parental level of education", hue="grades", data=d)
plt.show()
```

We observe from the above plot that most of the students scored B+, despite the fact that their parents had different levels of education. Hence, it can be concluded that the parental level of education does not play a huge role in the grade the student achieves.

#### GRADES VS ETHNICITY

```python
plt.figure(figsize=(15,8))
plt.title('Grades vs Ethinicity',fontsize=20)
sns.countplot(data=d, x="race/ethnicity", hue="grades")
plt.show()
```
